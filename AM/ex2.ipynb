```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

arquivo = 'EficiênciaVeículos.xlsx' #PRECISA ESTAR EM "ARQUIVOS"
df = pd.read_excel(arquivo)

#NOMES DAS COLUNAS NO ARQUIVO
coluna_peso = 'Peso (kg)'
coluna_eficiencia = 'Eficiência (km/l)'

X = df[[coluna_peso]].values
y = df[coluna_eficiencia].values

modelo = LinearRegression()
modelo.fit(X, y)

coeficiente_angular = modelo.coef_[0]
coeficiente_linear = modelo.intercept_

peso_alvo = np.array([[1300]])
eficiencia_predita = modelo.predict(peso_alvo)[0]

print(f"Coeficiente Angular: {coeficiente_angular:.4f}")
print(f"Coeficiente Linear: {coeficiente_linear:.4f}")
print(f"Eficiência prevista para 1300 kg: {eficiencia_predita:.2f} km/l")
```
