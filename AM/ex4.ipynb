```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# CARREGANDO O BANCO
dados_iris = datasets.load_iris()
caracteristicas = dados_iris.data
rotulos = dados_iris.target

# TREINO/TESTE
treino_x, teste_x, treino_y, teste_y = train_test_split(
    caracteristicas, rotulos, test_size=0.2, random_state=42
)

# PADRÃO
normalizador = StandardScaler()
treino_x_padrao = normalizador.fit_transform(treino_x)
teste_x_padrao = normalizador.transform(teste_x)

print("Dados preparados e normalizados com sucesso!")
```

```python
# TREINANDO MODELO POLINOMIAL
modelo_poly = SVC(kernel='poly', degree=3)
modelo_poly.fit(treino_x_padrao, treino_y)

# TESTE
previsao_poly = modelo_poly.predict(teste_x_padrao)
acerto_poly = accuracy_score(teste_y, previsao_poly)

print(f"Acurácia POLINOMIAL: {acerto_poly:.2%}")
```

```python
# TREINANDO MODELO RBF
modelo_rbf = SVC(kernel='rbf')
modelo_rbf.fit(treino_x_padrao, treino_y)

# TESTE
previsao_rbf = modelo_rbf.predict(teste_x_padrao)
acerto_rbf = accuracy_score(teste_y, previsao_rbf)

print(f"Acurácia RBF: {acerto_rbf:.2%}")
```
